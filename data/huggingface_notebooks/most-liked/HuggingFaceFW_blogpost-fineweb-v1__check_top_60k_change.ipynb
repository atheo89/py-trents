```python
import pandas as pd

df_43 = pd.read_csv('/home/gui/hf_dev/datatrove/blogpost/data/top_60k_urls_CC_MAIN_2021_43.csv').sort_values(by="Frequency", ascending=False)
df_49 = pd.read_csv('/home/gui/hf_dev/datatrove/blogpost/data/top_60k_urls_CC_MAIN_2021_49.csv').sort_values(by="Frequency", ascending=False)
```

```python
freqs_49 = {row[1]["URL"]: row[1]["Frequency"] for row in df_49.iterrows()}
```

```python
df_43['in_49'] = df_43.apply(lambda row: freqs_49.get(row["URL"], 0), axis=1)
df_43['change_to_49'] = df_43.apply(lambda row: freqs_49.get(row["URL"], 0) - row["Frequency"], axis=1)
```
