```python
# remove fully empty rows

df = df.dropna(how='all')
```

```python
# convert year to int

df['Year'] = df['Year'].fillna(0).astype(int)
```

```python
# convert datetime

df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
```

```python
# attendance missing values

df['Attendance'] = df['Attendance'].fillna(df['Attendance'].median())
```

```python
# create total goals column 

df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
```

```python
home = df['Home Team Name'].value_counts()
```

```python
away = df['Away Team Name'].value_counts()
```

```python
total_matches = home.add(away, fill_value=0).sort_values(ascending=False)
```

```python
total_matches.head(10).plot(kind='bar')
plt.title("Top 10 Teams by Matches Played")
plt.xlabel("Team")
plt.ylabel("Matches")
plt.show()
```

```python
goals_per_year = df.groupby('Year')['Total Goals'].sum()

goals_per_year.plot()
plt.title("Total Goals Scored Per Year")
plt.xlabel("Year")
plt.ylabel("Goals")
plt.show()
```

```python
attendance = df.groupby('Year')['Attendance'].mean()

attendance.plot()
plt.title("Average Attendance Per Year")
plt.xlabel("Year")
plt.ylabel("Attendance")
plt.show()
```

```python
top_matches = df.sort_values(by='Total Goals', ascending=False).head(10)

plt.barh(top_matches['Home Team Name'], top_matches['Total Goals'])
plt.title("Highest Goal Scoring Matches")
plt.xlabel("Goals")
plt.show()
```

```python
home_goals = df['Home Team Goals'].sum()
away_goals = df['Away Team Goals'].sum()

plt.bar(['Home Goals', 'Away Goals'], [home_goals, away_goals])
plt.title("Home vs Away Goals")
plt.show()
```

```python
df['City'].value_counts().head(10).plot(kind='bar')
plt.title("Top Cities Hosting Matches")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Between Features")
plt.show()
```

# Conclusion 

The exploratory data analysis of the FIFA World Cup matches dataset reveals important historical and performance-based insights about international football. The analysis shows that a few strong teams such as Brazil and Germany have participated in the highest number of matches, indicating their consistent qualification and dominance over different World Cup seasons. The total number of goals scored per year reflects the changing nature of the game, with modern tournaments generally showing higher goal counts due to improved playing strategies, fitness levels, and attacking styles.

The attendance trend demonstrates the growing global popularity of the FIFA World Cup, as the average number of spectators has increased significantly over time, especially in recent tournaments hosted in countries with large stadium capacities and advanced infrastructure. The comparison between home and away goals suggests the presence of a slight home-ground advantage, where teams playing in familiar conditions tend to perform better.

City and stadium analysis indicates that certain cities have hosted more matches, highlighting their importance as major football venues. The stage-wise match distribution confirms that the group stage contains the highest number of matches, as it involves all qualified teams, while the final stage consists of only one decisive match. High goal-scoring matches identified in the dataset represent the most competitive and entertaining games in World Cup history.

Finally, correlation analysis shows a strong relationship between half-time goals and full-time goals, which means early match performance often influences the final result, while attendance has little direct impact on the number of goals scored. Overall, this dataset provides valuable insights into team performance, match outcomes, audience engagement, and the evolution of football across different World Cup tournaments.
