# Who Decides the Target Column?
**Author:** Navjyot Architect
**Email:** navjyot.dataarchitect@proton.me

> **Big idea:** The dataset does **not** decide the target column.
>
>  **You decide the target column** based on the problem you want to solve.

---

## What is a Target Column?

In Machine Learning, the **target column** is simply:

> **The thing you want the model to predict.**

It is also called:

* label
* output
* y
* dependent variable

---

## Who Decides the Target Column?

###  The person who builds the ML project decides it.

That can be:

* you (as the learner)
* a company (business requirement)
* a client (problem statement)
* a research team (experiment goal)

A dataset is just a collection of columns.

> **Your goal decides what is the target.**

---

## The Easiest Way to Understand It (Plain English)

Imagine you have a dataset about **students**:

| hours_studied | attendance | previous_score | final_score |
| ------------- | ---------- | -------------- | ----------- |
| 2             | 80%        | 60             | 65          |
| 5             | 90%        | 70             | 78          |
| 8             | 95%        | 75             | 88          |

Now… which column is the target?

###  Answer: It depends on what you want.

If your goal is:

* **"Predict the final score"** → Target = `final_score`
* **"Predict attendance"** → Target = `attendance`
* **"Predict how many hours a student studied"** → Target = `hours_studied`

Same dataset.

Different target.

---

## The Most Important Rule

### The target must be something that makes sense to predict.

A good target should be:

* **useful** (solves a real question)
* **available in data** (you have the column)
* **not leaking information** (see next section)

---

## A Common Beginner Mistake: Data Leakage

Sometimes learners choose a target, but accidentally include a column that **gives away the answer**.

Example:

If you are predicting `fare`, and your dataset contains:

* `total_amount`
* `fare`
* `tax`
* `tolls`
