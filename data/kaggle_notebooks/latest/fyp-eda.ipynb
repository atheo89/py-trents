```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris, load_breast_cancer
from octolearn import AutoML
import json

print("✓ All dependencies imported successfully")
```

```python
# Load Iris dataset
data = load_iris(as_frame=True)
X = df.drop(columns={"Year"})
y = df['Year']

print(f"Dataset shape: {X.shape}")
print(f"\nFeatures: {X.columns.tolist()}")
print(f"\nData types:\n{X.dtypes}")
print(f"\nBasic statistics:\n{X.describe()}")
```

```python
automl = AutoML()
automl.fit(X, y)

print("✓ Model fitted successfully")
print(f"\nDataset Profile:")
print(f"  Hash: {automl.report().dataset_hash}")
print(f"  Task Type: {automl.report().task_type}")
```

```python
risk = automl.get_risk_score()

print(f"\n{'='*60}")
print(f"DATASET RISK SCORE: {risk['score']}/100")
print(f"CATEGORY: {risk['category']}")
print(f"{'='*60}")

if risk['factors']:
    print(f"\nRisk Factors Detected:")
    for factor_name, factor_desc in risk['factors'].items():
        print(f"  ⚠ {factor_desc}")
else:
    print("\n✓ No major risk factors detected")
```

```python
importance = automl.get_feature_importance()

#print("\nFeature Importance Scores:")
#print("-" * 45)
#for i, (feature, score) in enumerate(importance.items(), 1):
#    bar = '█' * int(score * 50)
#    print(f"{i:2}. {feature:25} {score:.4f} {bar}")
```

```python
suggestions = automl.get_preprocessing_suggestions()

print("\nPREPROCESSING STRATEGY:")
print("="*60)

for section, sug_list in suggestions.items():
    section_title = section.replace('_', ' ').upper()
    print(f"\n{section_title}:")
    print("-" * 60)
    
    if isinstance(sug_list, list):
        for suggestion in sug_list:
            print(f"  → {suggestion}")
    else:
        print(f"  → {sug_list}")
```

```python
print("\nGenerating comprehensive intelligence report...")
pdf_file = automl.generate_report()

print(f"\n✓ Report saved: {pdf_file}")
print(f"\nReport Contents:")
print("  📊 Risk Score Assessment")
print("  📈 Data Quality Analysis")
print("  📉 Feature Distributions")
print("  🔗 Correlation Heatmap")
print("  🎯 SHAP Feature Importance")
print("  🔧 Preprocessing Recommendations")
print("  💡 Strategic Automation Insights")
```
