# 🎓 Student Performance in Exams: Exploratory Data Analysis

### **Introduction**
The project involves the analysis of the performance of students in three major subjects: **Math, Reading, and Writing**. The aim is to gain insights into how different variables, including gender, ethnicity, the level of education of parents, types of lunch, and test preparation courses, affect the overall success of a student in academics.

### **Key Questions to Explore**
* **Predictors of Success:** What are the background variables (such as parental education or lunch type) that are the best predictors of high performance?
* **Preparation Impact:** Does the completion of a test preparation course lead to an overall improvement in performance in all three subjects?
* **Gender Trends:** Are there any gender-related trends in particular subjects (for instance, do men perform better in math than women)?
* **Subject Correlation:** To what extent are the three subjects correlated? (If a student is good at reading, is he also good at writing?)

### **Project Goal**
The aim of this Exploratory Data Analysis (EDA) is to graphically represent these relationships and gain key insights into the variables that play the most important part in the success of students.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Let's see the relation between the gender and the scores of the three exams

```python
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

The average scores for the men in math is greater than the women with 5 points but they are behind in the reading scores with 7-9 points

```python
df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']].mode()
```

Females are more frequent in this data set than males

```python
_, axes = plt.subplots(1, 3, figsize=(18, 5))
cols = ['math score', 'reading score', 'writing score']

for i, col in enumerate(cols):
    sns.histplot(df[col], kde=True, ax=axes[i])
    mean_val = df[col].mean()
    axes[i].axvline(mean_val, label=f'Mean: {mean_val:.2f}')
    axes[i].set_title(f'Distribution of {col.title()}')

plt.tight_layout()
plt.show()
```

```python
df[['math score', 'reading score', 'writing score']].mean()
```

The scores are normally distributed with near means from 66 to 68

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cat_cols = ['gender', 'race/ethnicity', 'test preparation course']

for i, col in enumerate(cat_cols):
    order = df[col].value_counts().index
    sns.countplot(data=df, x=col, ax=axes[i], palette='viridis', order=order, hue=col, legend=False)
    axes[i].set_title(f'Frequency of {col.title()}')
```

Now let's prepare the data to give grades to the average scores and make some analytics

```python
def Grade(Percentage):
    if (Percentage >= 88):return 'A'
    if (Percentage >= 72):return 'B'
    if (Percentage >= 60):return 'C'
    if (Percentage >= 50):return 'D'
    else: return 'F'
    
df["grade"] = df.apply(lambda x : Grade((x["math score"]+x["writing score"]+x["reading score"])/3.0), axis =1)
```

```python
df.head()
```

```python
grade_order = ["A","B","C","D","F"]
sns.countplot(
    data=df, 
    x='grade',
    hue = 'grade',
    order=grade_order,
    palette='deep',
)

plt.title('Grade Distribution by Race/Ethnicity', fontweight='bold')
plt.show()
```

They are normally distributed and deviated to the left

```python
sns.countplot(
    data=df, 
    x='grade',
    hue='race/ethnicity', 
    order=grade_order,
    palette='deep'
)

plt.title('Grade Distribution by Race/Ethnicity', fontweight='bold')
plt.show()
```
