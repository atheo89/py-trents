# library

```python
import torch
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
import torch.nn as nn
```

# Data

```python
data, feature = fetch_california_housing(return_X_y=True)
```

```python
X_train , X_valid , y_train , y_valid = train_test_split(data,feature,test_size=0.2)
```

## converting data into tensor

```python
X_train = torch.FloatTensor(X_train)
X_valid = torch.FloatTensor(X_valid)
```

```python
y_train = torch.FloatTensor(y_train).reshape(-1,1)
y_valid = torch.FloatTensor(y_valid).reshape(-1,1)
```

```python
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)
```

# standard scaling data

```python
X_train = (X_train-mean) / std
X_valid = (X_valid-mean) / std
```

```python
torch.manual_seed(8)
```

## creating the wights and the bais

```python
n_features = X_train.shape[1]
w = torch.rand((n_features,1), requires_grad=True)
b = torch.tensor(0., requires_grad=True)
```

# training the model

```python
learning_rate = 0.04
n_epochs = 100
for epoch in range(n_epochs):
  y_pred = X_train @ w + b
  loss = ((y_pred - y_train)**2).mean()
  loss.backward()
  with torch.no_grad():
    b -= learning_rate * b.grad
    w -= learning_rate * w.grad
    b.grad.zero_()
    w.grad.zero_()

  print(f"Epoch {epoch + 1}/{n_epochs}, Loss: {loss.item()}")
```

## evaluating the model

```python
def r2_score(y_true,y_pred):
  r = 1 - (((y_true - y_pred) **2).sum())  / (((y_true - y_true.mean()) **2).sum())
  return r
```

```python
with torch.no_grad():
  y_pred = X_train @ w +b
```

```python
score = r2_score(y_train,y_pred)
print(f"train r2 score :{score}")
```

```python
with torch.no_grad():
  y_valid_pred = X_valid @ w +b
```

```python
score = r2_score(y_valid,y_valid_pred)

print(f"valid r2 score :{score}")
```
