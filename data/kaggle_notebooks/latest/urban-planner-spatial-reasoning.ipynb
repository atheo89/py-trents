```python
import kaggle_benchmarks as kbench

@kbench.task(name="multi_modal_transport_plan")
def multi_modal_transport_plan(llm, judge_llm):
    prompt = """
    SCENARIO: A 6-lane highway connects to a 2-lane residential road at [0,1]. 
    Traffic flow is at 5%. You are FORBIDDEN from adding more car lanes.
    
    TASK: Suggest a 3-step 'Multi-Modal' plan to solve the jam. 
    Include: A Metro/Bus solution, a Bike/Pedestrian solution, and a Zoning policy change.
    """
    response = llm.prompt(prompt)

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=[
            "Does it suggest Public Transport (Metro/Bus)?",
            "Does it mention Active Transport (Bikes/Walking)?",
            "Does it suggest a Policy/Zoning change?",
            "Did it avoid adding more car lanes?",
        ],
        response_text=response,
        judge_llm=judge_llm
    )

    for result in assessment.results:
        kbench.assertions.assert_true(
            result.passed,
            expectation=f"Criterion '{result.criterion}' failed: {result.reason}"
        )

multi_modal_transport_plan.run(llm=kbench.llm, judge_llm=kbench.judge_llm)
```
