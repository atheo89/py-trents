```python
import requests
import json
import uuid  # specific library to generate unique IDs
from datetime import datetime

# --- CONFIGURATION ---
# 1. Open Developer Tools (F12) -> Network -> Click 'chat' request -> Headers.
# 2. Copy the entire value starting with "eyJ..." from the 'Authorization' header.
AUTH_TOKEN = "Bearer YOUR_VERY_LONG_TOKEN_HERE" 

# Target Endpoint
URL = "https://cibc-ai-backend-app.pase-cane-147108-prod-01.appserviceenvironment.net/api/chat"

def run_scraper():
    # Generate fresh IDs to avoid conflicts with previous sessions
    current_conversation_id = str(uuid.uuid4())
    current_chat_session_id = str(uuid.uuid4())
    
    # Headers exactly matching the screenshots
    headers = {
        "Authority": "cibc-ai-backend-app.pase-cane-147108-prod-01.appserviceenvironment.net",
        "Accept": "application/json, text/plain, */*",
        "Accept-Encoding": "gzip, deflate, br, zstd",
        "Accept-Language": "en-US,en;q=0.9",
        "Authorization": AUTH_TOKEN,  # CRITICAL: Must be fresh
        "Chatsessionid": current_chat_session_id, # Dynamic ID
        "Content-Type": "application/json",
        "Origin": "https://cai.cibc.com",
        "Referer": "https://cai.cibc.com/",
        "Sec-Ch-Ua": '"Google Chrome";v="143", "Chromium";v="143", "Not A(Brand";v="24"',
        "Sec-Ch-Ua-Mobile": "?0",
        "Sec-Ch-Ua-Platform": '"Windows"',
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36"
    }

    # Payload matching the structure in 49561.jpg
    payload = {
        "chatmessages": [
            {
                "content": "Hello, what are the current mortgage rates?", # Your prompt here
                "role": "user",
                "fileKeys": [],
                "email_ids": [],
                "OneDriveFiles": [],
                "calendar_event_ids": []
            }
        ],
        "conversation_id": current_conversation_id,
        "is_marketplace_recipe": False,
        "is_submission_recipe": False,
        "language_code": "en",
        "model": "gpt-4.1",
        "recipe_id": 0,
        "selected_region": None,
        "stream": False,
        "temperature": "balanced",
        "time_zone": "Asia/Calcutta"
    }

    print(f"[-] Sending request with Conversation ID: {current_conversation_id}")

    try:
        response = requests.post(URL, headers=headers, json=payload, timeout=30)
        
        # --- RESPONSE HANDLING ---
        if response.status_code == 200:
            data = response.json()
            
            # Extracting the actual answer based on screenshot 49562.jpg
            ai_reply = data.get("response", "No response text found.")
            query_id = data.get("query_id")
            
            print("\n[+] SUCCESS!")
            print("-" * 40)
            print(f"AI Response: {ai_reply}")
            print("-" * 40)
            print(f"Query ID: {query_id}")
            
            # Optional: Save full log for debugging
            # with open("response_log.json", "w") as f:
            #     json.dump(data, f, indent=4)
                
        elif response.status_code == 401:
            print("\n[!] ERROR 401: Unauthorized.")
            print("    Your 'Authorization' token is invalid or expired.")
            print("    Please get a new token from the browser Network tab.")
            
        elif response.status_code == 403:
            print("\n[!] ERROR 403: Forbidden.")
            print("    WAF (Firewall) blocked the request or you lack permission.")
            
        else:
            print(f"\n[!] Unexpected Error: {response.status_code}")
            print(response.text)

    except requests.exceptions.ConnectionError:
        print("\n[!] Connection Error. You might need to be on the corporate VPN.")
    except Exception as e:
        print(f"\n[!] Script crashed: {str(e)}")

if __name__ == "__main__":
    run_scraper()
```
