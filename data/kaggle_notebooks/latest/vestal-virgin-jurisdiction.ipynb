```python
import kaggle_benchmarks as kbench

@kbench.task(name="vestal_virgin_jurisdiction")
def vestal_virgin_jurisdiction(llm):
    prompt = """
    If a Vestal Virgin was legally emancipated from paternal authority (patria potestas)
    and possessed independent legal status (sui iuris), under what authority could she
    be punished — and by whom — if accused of violating her vow of chastity?
    """
    response = llm.prompt(prompt)

    criteria = [
        "The response correctly identifies that the Vestal remained under the religious authority (auctoritas sacra) of the Pontifex Maximus.",
        "The response states that jurisdiction was religious, not civil, and the investigation was conducted by the Pontifex Maximus.",
        "The response correctly identifies the College of Pontiffs (pontifices) as the body that determined guilt.",
        "The response accurately describes the penalty for the Vestal as being buried alive in the Campus Sceleratus.",
        "The response mentions that the male accomplice was scourged to death.",
        "The response explains that the punishment was framed as a ritual expiation to avoid direct bloodshed and remove religious pollution.",
    ]

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=criteria,
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM should have provided an assessment."
        )

    for result in assessment.results:
        kbench.assertions.assert_true(
            result.passed,
            expectation=f"Criterion '{result.criterion}' failed: {result.reason}"
        )

vestal_virgin_jurisdiction.run(kbench.llm)
```
