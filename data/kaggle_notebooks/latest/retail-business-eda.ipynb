# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing data

```python
df = pd.read_csv("/kaggle/input/retail-business-analytics-dataset-10k-orders/Business_Analytics_Dataset_10000_Rows.csv")
df.head()
```

# EDA (Exploratory Data Analysis)

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
df['Order_Month'] = df['Order_Date'].dt.month
```

```python
df['Order_Year'] = df['Order_Date'].dt.year
```

```python
df.columns
```

```python
df.info()
```

```python
sns.scatterplot(x='Region', y='Revenue', data = df)
plt.title('Region vs Revenue')
```

```python
region_data = df.groupby('Region')['Revenue'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Revenue', data=region_data)

plt.title('Region vs Revenue')
plt.xlabel('Region')
plt.ylabel('Total Revenue')

plt.show()
```

```python
region_data = df.groupby('Region')['Profit'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Profit', data=region_data)

plt.title('Region vs Profit')
plt.xlabel('Region')
plt.ylabel('Profit')

plt.show()
```

```python
sns.scatterplot(x='Discount_Rate', y='Profit', data = df)
plt.title('Profit on basis of Dicount_Rate')

plt.title('Profit vs Discount Rate')
plt.show()
```

```python
sns.barplot(x='Product_Category', y='Revenue', data = df, estimator=sum)
plt.title('Category vs Revenue')
```

```python
monthly_data = df.groupby('Order_Month')['Revenue'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(x='Order_Month', y='Revenue', data=monthly_data)

plt.title('Month wise Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x='Cost', y='Profit', data=df)

plt.title('Cost vs Profit')
plt.xlabel('Cost')
plt.ylabel('Profit')

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Profit'], kde=True)
plt.title('Profit Distribution Highlighting Margin Variability')
plt.show()
```

```python
category_sales = df.groupby('Product_Category')['Profit'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.barplot(x='Product_Category', y='Profit', data=category_sales,)
plt.title('Profitability Comparison by Product Category')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Customer_Segment', y='Revenue', data=df, estimator=sum)
plt.title('Total Revenue Contribution by Customer Segment')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Quantity', y='Revenue', data=df)
plt.title('Relationship Between Quantity Sold and Revenue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Cost', y='Revenue', data=df)
plt.title('Cost vs Revenue Correlation Analysis')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Quantity','Unit_Price','Discount_Rate','Revenue','Cost','Profit']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Key Financial Metrics')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Payment_Method', data=df)
plt.title('Payment Method Usage Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product_Category', y='Quantity', data=df, estimator=sum)
plt.title('Total Units Sold by Product Category')
plt.show()
```
