# **DATA PREPROCESSING:**

```python
df.columns
```

```python
df.head()
```

```python
# Encode the "class" feature, the model will predict if a mushroom is edible(1) or not (0)
df["class"] = df["class"].map({
    "e": 1, 
    "p": 0
})

df.head()
```

```python
# Encode the "cape-shape" feature
encoded_cap_shape = pd.get_dummies(df["cap-shape"]).astype(int)

# Since most of the features have the same categorical values, we will rename the columns to avoid confusion
encoded_cap_shape.columns = [f"cap_shape_{col}" for col in encoded_cap_shape.columns]

# Merge the encoded feature and the dataframe
df = pd.concat([df, encoded_cap_shape], axis=1)

# Drop the "cap-shape" column
df = df.drop(["cap-shape"], axis=1)

df.head()
```

```python
# Encode the "cape-surface" feature
encoded_cap_surface = pd.get_dummies(df["cap-surface"]).astype(int)

# Rename the columns to avoid confusion
encoded_cap_surface.columns = [f"cap_surface_{col}" for col in encoded_cap_surface.columns]

# Merge the encoded feature and the dataframe
df = pd.concat([df, encoded_cap_surface], axis=1)

# Drop the "cap-surface" column
df = df.drop(["cap-surface"], axis=1)

df.head()
```

# **DATA SPLITTING:**

```python
df.shape
```

```python
# Get x and y
X = df.drop(["class"], axis=1)
y = df["class"]
```

```python
X.head()
```

```python
y
```

```python
# Training and testing split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

```python
X_train.shape, X_test.shape
```

# **MODEL TRAINING AND EVALUATION:**

```python
# Model implementation and training
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(random_state=0)
forest.fit(X_train, y_train)
```

```python
# Model evaluation
forest.score(X_train, y_train), forest.score(X_test, y_test)
```
