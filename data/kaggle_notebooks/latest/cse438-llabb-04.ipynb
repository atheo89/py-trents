```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image (1).png', cv2.IMREAD_GRAYSCALE) 
```

```python
def add_salt_pepper_noise(image, salt_prob=0.02, pepper_prob=0.02):
    noisy = image.copy().astype(np.float32)
    total_pixels = image.size
    
    # Salt 
    num_salt = np.ceil(salt_prob * total_pixels).astype(int)
    coords = [np.random.randint(0, i-1, num_salt) for i in image.shape]
    noisy[coords[0], coords[1]] = 255
    
    # Pepper 
    num_pepper = np.ceil(pepper_prob * total_pixels).astype(int)
    coords = [np.random.randint(0, i-1, num_pepper) for i in image.shape]
    noisy[coords[0], coords[1]] = 0
    
    return noisy.astype(np.uint8)

noisy_img = add_salt_pepper_noise(img, salt_prob=0.03, pepper_prob=0.03)

kernel_size = 3   

# Max filter 
max_filtered = cv2.dilate(noisy_img, np.ones((kernel_size, kernel_size), np.uint8))

# Min filter 
min_filtered = cv2.erode(max_filtered, np.ones((kernel_size, kernel_size), np.uint8))
                         
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("With Salt & Pepper Noise")
plt.imshow(noisy_img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("After Min + Max Filter")
plt.imshow(min_filtered, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

img = cv2.imread('/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image (2).png', cv2.IMREAD_GRAYSCALE)

def add_gaussian_noise(image, mean=0, sigma=25):
    row, col = image.shape
    gauss = np.random.normal(mean, sigma, (row, col))
    noisy = image.astype(np.float32) + gauss
    noisy = np.clip(noisy, 0, 255)   
    return noisy.astype(np.uint8)

noisy_img = add_gaussian_noise(img, mean=0, sigma=25)   

filtered_img = cv2.GaussianBlur(noisy_img, ksize=(5, 5), sigmaX=0)


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("With Gaussian Noise")
plt.imshow(noisy_img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("After Gaussian Filter")
plt.imshow(filtered_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img_path = 'knee.jpg'          
img = cv2.imread('/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image (3).png', cv2.IMREAD_GRAYSCALE)

def add_salt_pepper_noise(image, salt_prob=0.03, pepper_prob=0.03):
    noisy = image.copy().astype(np.float32)
    total = image.size
    
    # Salt 
    num_salt = np.ceil(salt_prob * total).astype(int)
    coords = [np.random.randint(0, i-1, num_salt) for i in image.shape]
    noisy[coords[0], coords[1]] = 255
    
    # Pepper 
    num_pepper = np.ceil(pepper_prob * total).astype(int)
    coords = [np.random.randint(0, i-1, num_pepper) for i in image.shape]
    noisy[coords[0], coords[1]] = 0
    
    return np.clip(noisy, 0, 255).astype(np.uint8)

noisy_img = add_salt_pepper_noise(img, salt_prob=0.04, pepper_prob=0.04)


kernel_size = 5   


box_filtered = cv2.boxFilter(noisy_img, ddepth=-1, ksize=(kernel_size, kernel_size), normalize=False)


avg_filtered = cv2.blur(noisy_img, ksize=(kernel_size, kernel_size))


median_filtered = cv2.medianBlur(noisy_img, ksize=kernel_size)


plt.figure(figsize=(20, 8))

titles = ["Original", "Noisy (Salt & Pepper)", "Box Filter", "Average Filter", "Median Filter"]
images = [img, noisy_img, box_filtered, avg_filtered, median_filtered]

for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(box_filtered, cmap='gray')
plt.title("Box Filter")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(avg_filtered, cmap='gray')
plt.title("Average Filter")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter (Best)")
plt.axis('off')

plt.suptitle("Comparison of 3 Filtering Techniques", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
img_path = '/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image (6).png'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  

if img is None:
    print("Image not found! Please check the file name/path.")
exit()

gaussian_kernel = (9, 9)   
sigma = 3.0                 
amount_unsharp = 1.5         
amount_highboost = 2.5       

blurred = cv2.GaussianBlur(img, gaussian_kernel, sigma)

mask_unsharp = img.astype(np.float32) - blurred.astype(np.float32)
unsharp_img = img.astype(np.float32) + amount_unsharp * mask_unsharp
unsharp_img = np.clip(unsharp_img, 0, 255).astype(np.uint8)

highboost_img = img.astype(np.float32) + amount_highboost * mask_unsharp
highboost_img = np.clip(highboost_img, 0, 255).astype(np.uint8)

plt.figure(figsize=(15, 8))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original MRI")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(blurred, cmap='gray')
plt.title("Blurred (Base)")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(unsharp_img, cmap='gray')
plt.title("g) Unsharp Masking")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(highboost_img, cmap='gray')
plt.title("h) High Boost Filtering")
plt.axis('off')

plt.tight_layout()
plt.show()

crop_original = img[100:300, 200:400]
crop_unsharp  = unsharp_img[100:300, 200:400]
crop_highboost = highboost_img[100:300, 200:400]

plt.figure(figsize=(12, 4))
plt.subplot(131); plt.imshow(crop_original, cmap='gray'); plt.title("Original Crop"); plt.axis('off')
plt.subplot(132); plt.imshow(crop_unsharp, cmap='gray'); plt.title("Unsharp Crop"); plt.axis('off')
plt.subplot(133); plt.imshow(crop_highboost, cmap='gray'); plt.title("High Boost Crop"); plt.axis('off')
plt.suptitle("Zoomed Comparison (Edges & Details)", fontsize=10)
plt.show()
```

```python
img_path = '/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image (7).png'  
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  

if img is None:
    print("Image not found! Please check the file name/path.")
exit()

laplacian = cv2.Laplacian(img, cv2.CV_64F)


laplacian_abs = np.uint8(np.absolute(laplacian))

c = 1.0          
sharpened = img.astype(np.float64) + c * laplacian

sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)

kernel = np.array([[0, 1, 0],
                   [1, -4, 1],
                   [0, 1, 0]])
lap_custom = cv2.filter2D(img, cv2.CV_64F, kernel)
sharpened_custom = np.clip(img.astype(np.float64) + c * lap_custom, 0, 255).astype(np.uint8)

plt.figure(figsize=(8, 8))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Spine Image")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(laplacian_abs, cmap='gray')
plt.title("Laplacian (Absolute)")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened (c=1.0)")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(sharpened_custom, cmap='gray')
plt.title("Sharpened (Custom Kernel)")
plt.axis('off')

plt.tight_layout()
plt.show()

crop_row_start, crop_row_end = 100, 300
crop_col_start, crop_col_end = 150, 350

plt.figure(figsize=(8, 4))
plt.subplot(131); plt.imshow(img[crop_row_start:crop_row_end, crop_col_start:crop_col_end], cmap='gray'); plt.title("Original Crop"); plt.axis('off')
plt.subplot(132); plt.imshow(sharpened[crop_row_start:crop_row_end, crop_col_start:crop_col_end], cmap='gray'); plt.title("Sharpened Crop"); plt.axis('off')
plt.subplot(133); plt.imshow(sharpened_custom[crop_row_start:crop_row_end, crop_col_start:crop_col_end], cmap='gray'); plt.title("Custom Kernel Crop"); plt.axis('off')
plt.suptitle("Zoomed-in Edges Comparison", fontsize=14)
plt.show()
```

```python
img_path = '/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image(89).png'   
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
   print("Image not found! Please check the file name/path.")
exit()

# 1. Roberts Cr
roberts_x = cv2.filter2D(img, cv2.CV_64F, np.array([[1, 0], [0, -1]]))
roberts_y = cv2.filter2D(img, cv2.CV_64F, np.array([[0, 1], [-1, 0]]))
roberts = np.sqrt(roberts_x**2 + roberts_y**2)
roberts = np.uint8(np.clip(roberts, 0, 255))   

# 2. Sobel
sobel_x = cv2.Sobel(img, cv2.CV_64F, dx=1, dy=0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, dx=0, dy=1, ksize=3)
sobel = np.sqrt(sobel_x**2 + sobel_y**2)
sobel = np.uint8(np.clip(sobel, 0, 255))

# 3. Prewitt
prewitt_kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
prewitt_x = cv2.filter2D(img, cv2.CV_64F, prewitt_kernel_x)
prewitt_y = cv2.filter2D(img, cv2.CV_64F, prewitt_kernel_y)
prewitt = np.sqrt(prewitt_x**2 + prewitt_y**2)
prewitt = np.uint8(np.clip(prewitt, 0, 255))

plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original CT Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(roberts, cmap='gray')
plt.title("Roberts Cross Edge Detection")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(sobel, cmap='gray')
plt.title("Sobel Edge Detection")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(prewitt, cmap='gray')
plt.title("Prewitt Edge Detection")
plt.axis('off')

plt.tight_layout()
plt.show()

crop = slice(150, 350), slice(150, 350)  

plt.figure(figsize=(8, 4))
plt.subplot(141); plt.imshow(img[crop], cmap='gray'); plt.title("Original"); plt.axis('off')
plt.subplot(142); plt.imshow(roberts[crop], cmap='gray'); plt.title("Roberts"); plt.axis('off')
plt.subplot(143); plt.imshow(sobel[crop], cmap='gray'); plt.title("Sobel"); plt.axis('off')
plt.subplot(144); plt.imshow(prewitt[crop], cmap='gray'); plt.title("Prewitt"); plt.axis('off')
plt.suptitle("Zoomed-in Edge Comparison", fontsize=10)
plt.show()
```

```python
img_path = '/kaggle/input/datasets/meenmoyafrinafrin/cse-438-lab-04/image(89).png'   
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Image not found! Please check the file name/path.")
exit()

# Parameters 
blur_ksize = (5, 5)
sigma = 1.5
unsharp_amount = 1.5       
highboost_k = 2.5          
lap_c = 1.0                
canny_low = 80
canny_high = 180

# Blurred base for sharpening methods
blurred = cv2.GaussianBlur(img, blur_ksize, sigma)

# 1. Unsharp Maskin
mask = img.astype(float) - blurred.astype(float)
unsharp = np.clip(img.astype(float) + unsharp_amount * mask, 0, 255).astype(np.uint8)

# 2. High Boost Filtering
highboost = np.clip(img.astype(float) + highboost_k * mask, 0, 255).astype(np.uint8)

# 3. Laplacian Sharpening
laplacian = cv2.Laplacian(img, cv2.CV_64F)
sharpen_lap = np.clip(img.astype(float) + lap_c * laplacian, 0, 255).astype(np.uint8)

# 4. Roberts-cross
rob_x = cv2.filter2D(img, cv2.CV_64F, np.array([[1, 0], [0, -1]]))
rob_y = cv2.filter2D(img, cv2.CV_64F, np.array([[0, 1], [-1, 0]]))
roberts = np.sqrt(rob_x**2 + rob_y**2)
roberts = np.uint8(np.clip(roberts, 0, 255))

# 5. Sobel
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = np.sqrt(sobelx**2 + sobely**2)
sobel = np.uint8(np.clip(sobel, 0, 255))

# 6. Prewitt
pre_x = cv2.filter2D(img, cv2.CV_64F, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
pre_y = cv2.filter2D(img, cv2.CV_64F, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
prewitt = np.sqrt(pre_x**2 + pre_y**2)
prewitt = np.uint8(np.clip(prewitt, 0, 255))

# 7. Canny
canny = cv2.Canny(img, canny_low, canny_high)

plt.figure(figsize=(10, 8))

titles = [
    "Original", "Unsharp Masking", "High Boost", "Laplacian Sharpen",
    "Roberts-cross", "Sobel", "Prewitt", "Canny"
]
results = [img, unsharp, highboost, sharpen_lap, roberts, sobel, prewitt, canny]

for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(results[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Performance Comparison of Sharpening & Edge Detection Methods", fontsize=16)
plt.show()

crop_row = slice(80, 280)
crop_col = slice(120, 320)

plt.figure(figsize=(10, 8))
for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(results[i][crop_row, crop_col], cmap='gray')
    plt.title(titles[i] + " (Zoom)")
    plt.axis('off')

plt.tight_layout()
plt.suptitle("Zoomed Comparison - Edges & Details", fontsize=8)
plt.show()
```
