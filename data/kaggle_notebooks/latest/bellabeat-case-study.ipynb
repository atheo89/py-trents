## Data sources
The dataset has been collected and displayed through 18 csv files for public non-commercial use directly on Kaggle by the author, Möbius (https://www.kaggle.com/datasets/arashnic/fitbit). It is a public dataset stored in the wide format. In terms of the ROCCC data guidelines -  whether the data is reliable, original, comprehensive, current, and cited - it does meet all criteria. While it does meet sufficient ROCCC qualifications, it is important to outline potential limitations of problems with the given data, which is done below.

Importantly, there are various drawbacks related to this specific dataset that are noteworthy. Primarily, there is no specific demographic information given regarding the anonymous users included in this dataset. This anonymity does provide the opportunity for ensured non-bias analysis regarding specific users and their habits. On the other end, however, this anonymity and lack of information holds the data-driven solutions back in regards to further analysis using the additional information. For example, the data does not specify any user's age, geographical location, working status / lifestyle, or medical conditions. Another potential limitation is the currency status of the data. It will soon be a decade old, and the smart device industry has grown immensely, so more current or up-to-date data may provide a more efficient or accurate reflection of smart device trends.

This dataset was recommended directly based off the needs and desires of the stakeholders who aim to answer their business inquiries through this smart device data. ## Data Analysis

Converted minuteSleep to hourSleep and grouped by Ids to provide total sleep time per user per day. 
```
CREATE OR REPLACE TABLE minuteSleep_merged.hourSleep AS SELECT
  Id,
  SUM(value) AS total_value
FROM minuteSleep_merged.minuteSleep_cleaned
GROUP BY Id;
```
Then, we analyzed the minuteSleep information to provide insights based on Fitbit user's sleep schedule.
