```python
from deep_translator import GoogleTranslator
import nltk

nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer

SIA = SentimentIntensityAnalyzer()
```

```python
translated_text = GoogleTranslator(source='ar', target='en').translate('كيف حالك')

print(translated_text)
```

```python
def T(text):
    return GoogleTranslator(source='ar', target='en').translate(text)
```

```python
T('عامل ايه')
```

```python
T('وكابو، الواقعة على بعد أقل من 15 ميلاً (24 كيلومتراً) شرق ميناء بيرل هاربر في صباح 7 ديسمبر 1941، تكبدت الولايات المتحدة أكبر الخسائر العسكرية في تاريخها على الإطلاق، واضطرت لدخول الحرب العالمية الثانية')
```

```python
Values = list(SIA.polarity_scores(T('فيلم زي الزفت بصراحة')).values())

print(f'Value for Negative is  : {Values[0]}')
print(f'Value for Neutral is   : {Values[1]}')
print(f'Value for Positive is  : {Values[2]}')
print(f'Value for Compound is  : {Values[3]}')
```

```python
def TranslateAll(T):
    for i, k in GoogleTranslator().get_supported_languages(as_dict=True).items():
        print(f'For Language {i}, translation is ({GoogleTranslator(source="ar", target=k).translate(T)})\n' + '.'*25 + '\n')

TranslateAll('كيف حالك')
```
