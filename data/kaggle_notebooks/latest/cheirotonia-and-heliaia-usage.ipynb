```python
import kaggle_benchmarks as kbench

@kbench.task(name="cheirotonia_and_heliaia_usage")
def check_cheirotonia_usage(llm):
    """
    Checks the model's knowledge about the use of cheirotonia (show of hands)
    in ancient Athenian institutions, specifically its non-use in Heliaia verdicts.
    """
    question = "In ancient Athens, what was the relationship between cheirotonia (χειροτονία, a show of hands) and the Heliaia (the supreme court)?"
    response = llm.prompt(question)

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=[
            "The response must correctly state that cheirotonia was NOT used for verdicts in the Heliaia.",
            "The response should mention that the Heliaia used a different method for verdicts, typically secret ballots (psephoi).",
            "The response should correctly identify the Ekklesia (Assembly) as a primary body where cheirotonia was used.",
            "The response may note that cheirotonia could be used for some procedural matters, but clarifies it was not for the final court verdict."
        ],
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM failed to provide an assessment."
        )
    else:
        for result in assessment.results:
            kbench.assertions.assert_true(
                result.passed,
                expectation=f"Criterion '{result.criterion}' failed: {result.reason}"
            )

check_cheirotonia_usage.run(kbench.llm)
```
