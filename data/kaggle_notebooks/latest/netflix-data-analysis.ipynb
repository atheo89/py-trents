# Netflix Data Analysis: Movies, TV Shows, Genres, and Ratings

## Introduction
This notebook explores Netflix content including movies, TV shows, genres, ratings, and duration. 
The goal is to derive insights using Python, Pandas, and Matplotlib.
We will analyze trends, top countries, popular genres, rating distribution, and movie durations.


## 1. Data Loading and Overview
We load the Netflix dataset and check its structure, number of rows and columns, data types, and missing values.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



df=pd.read_csv("/kaggle/input/datasets/shivamb/netflix-shows/netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

## 2. Data Cleaning
- Handle missing values in 'country', 'rating', 'date_added', 'duration'.
- Drop duplicates if any.
- Convert 'date_added' to datetime object.
- Convert 'duration' to numeric 'duration_num'.


```python
df1=df.copy()
df1.isnull().sum()
```

```python
df1["country"]=df1["country"].fillna("Unknown")
df1=df1.dropna(subset=['date_added'])
df1=df1.dropna(subset=['rating'])
df1=df1.dropna(subset=['duration'])
df1.isnull().sum()
```

```python
df1.duplicated().sum()
```

```python
df1.dtypes
```

```python
df1['date_added'] = pd.to_datetime(df1['date_added'], errors='coerce')
df1 = df1.dropna(subset=['date_added'])
df1['duration_num'] = df1['duration'].str.extract('(\d+)').astype(int)
df1.dtypes
```

## 3. Movies vs TV Shows
This chart shows the number of movies and TV shows available on Netflix.


```python
df_type_count = df1.groupby("type")["show_id"].count()

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(df_type_count.index, df_type_count.values, color=['skyblue', 'salmon'])

plt.title("Number of Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

## 4. Top Countries by Netflix Content
We visualize the top 10 countries producing Netflix content.


```python
country_counts = df1.groupby('country')['show_id'].count()
top_countries = country_counts.sort_values(ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(top_countries.index, top_countries.values, color='mediumseagreen')

plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Titles", fontsize=12)
plt.title("Top 10 Countries by Netflix Content", fontsize=16)
plt.xticks(rotation=45)  # Rotate x labels for readability

plt.show()
```

## 5. Content Added Over the Years
Shows how Netflix content has grown over time.


```python
df1['year_added'] = df1['date_added'].dt.year
shows_per_year=df1.groupby('year_added')['show_id'].count()

#plot line chart
plt.figure(figsize=(12,6))
plt.plot(shows_per_year.index,shows_per_year.values,marker='o')

plt.title("Netflix content added over the years",fontsize=16)
plt.xlabel("Year Added",fontsize=12)
plt.ylabel("No of titles",fontsize=12)
plt.grid(True)
plt.xticks(shows_per_year.index, rotation=45)  # Rotate years for readability

plt.show()
```

## 6. Top 10 Genres on Netflix
Analyzing which genres are most common on Netflix.


```python
all_genres=df1["listed_in"].str.split(",").explode()
top_genres = all_genres.value_counts().head(10)

plt.figure(figsize=(12,6))
plt.bar(top_genres.index, top_genres.values, color='blue')

plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Titles", fontsize=12)
plt.title("Top 10 Netflix Genres", fontsize=16)
plt.xticks(rotation=45)  # Rotate for readability

plt.show()
```

## 7. Rating Distribution
Analyzing how Netflix content is rated (TV-MA, PG-13, etc.)


```python
ratings_counts=df1.groupby("rating")["show_id"].count().sort_values(ascending=False)
plt.figure(figsize=(12,6))
plt.bar(ratings_counts.index,ratings_counts.values)
plt.xlabel("Rating",fontsize=12)
plt.ylabel("Number of titles",fontsize=12)
plt.title("Netflix Content by Rating",fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

## 8. Movie Duration Distribution
Analyzing the distribution of movie durations in minutes.


```python
movies = df1[df1['type'] == 'Movie']
plt.figure(figsize=(12,6))
plt.hist(movies['duration_num'], bins=20, color='skyblue', edgecolor='black')

plt.xlabel("Duration (minutes)", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.title("Distribution of Movie Durations on Netflix", fontsize=16)
plt.grid(True)
plt.show()
```

## 9. Conclusion
- Netflix has more movies than TV shows.
- USA, India, and UK produce the most content.
- International movies,Drama and Comedy dominate genres.
- Most content is rated TV-MA and PG-13.
- Movies are mostly between 80-120 minutes.
