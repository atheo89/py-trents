# String Methods Lab Assignment
### Section 1: Basic String Operations

* **Question 1:** Concatenate your first name and last name with a space in between.

```python
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Full Name:", full_name)
```

* **Question 2:** Repeat the string "Python" 5 times.

```python
repeated_str = "Python" * 5
print("Repeated String:", repeated_str)
```

## Section 2: String Methods
* **Question 3:** Convert the string "Python Programming" to uppercase.


```python
str2 = "Python Programming"
upper_str2 = str2.upper()
print("Uppercase:", upper_str2)
```

* **Question 4:** Remove leading and trailing spaces from the string " Data Science ".

```python
str3 = "  Data Science  "
stripped_str3 = str3.strip()
print("Stripped:", stripped_str3)
```

## Section 3: Advanced String Methods

* **Question 5:** Replace "Data" with "AI" in the string "Data Science".


```python
str4 = "Data Science"
replaced_str4 = str4.replace("Data", "AI")
print("Replaced:", replaced_str4)

```

* **Question 6:** Split the string "Machine Learning" into words and then join them with a hyphen.

```python
str5 = "Machine Learning"
words5 = str5.split()
joined_str5 = "-".join(words5)
print("Joined with hyphen:", joined_str5)
```

## Section 4: String Checks

* **Question 7:** Check if the string "2025" is numeric.


```python
str6 = "2025"
is_numeric = str6.isnumeric()
print("Is numeric:", is_numeric)
```

* **Question 8:** Check if the string "Python" contains only alphabetic characters.

```python
str7 = "Python"
is_alpha = str7.isalpha()
print("Is alpha:", is_alpha)
```

* **Question 9:** Check if the string "AI2025" contains only alphanumeric characters.

**Note:** **Alphanumeric** refers to a combination of alphabetic and numeric characters. In other words, it includes:


* Alphabetic characters: Letters from A to Z (both uppercase and lowercase).
* Numeric characters: Digits from 0 to 9.


```python
str8 = "AI2025"
is_alnum = str8.isalnum()
print("Is alphanumeric:", is_alnum)
```

* **Question 10:** Check if the string "AI-2025" contains only alphanumeric characters.

**Note:** **Alphanumeric** refers to a combination of alphabetic and numeric characters. In other words, it includes:


* Alphabetic characters: Letters from A to Z (both uppercase and lowercase).
* Numeric characters: Digits from 0 to 9.

```python
str9 = "AI-2025"
is_alnum = str9.isalnum()
print("Is alphanumeric:", is_alnum)
```
