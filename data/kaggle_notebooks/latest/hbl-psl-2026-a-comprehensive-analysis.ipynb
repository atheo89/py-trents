## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Enable inline plotting
%matplotlib inline
```

## Load Dataset

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/hbl-psl-2026-players-prices/hbl_psl_2026_players.csv')

# Display first few rows
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

## Price Analysis

```python
# Price statistics
print("Price Statistics (PKR):")
print(f"Total Players: {len(df)}")
print(f"Highest Price: PKR {df['Price (PKR)'].max():,}")
print(f"Lowest Price: PKR {df['Price (PKR)'].min():,}")
print(f"Average Price: PKR {df['Price (PKR)'].mean():,.2f}")
print(f"Median Price: PKR {df['Price (PKR)'].median():,.2f}")
print(f"Total Value: PKR {df['Price (PKR)'].sum():,}")
```

## Top 10 Most Expensive Players

```python
# Top 10 most expensive players
top_10 = df.nlargest(10, 'Price (PKR)')[['Player Name (English)', 'Price (PKR)', 'Price (Urdu)']]

print("Top 10 Most Expensive Players:\n")
for idx, row in top_10.iterrows():
    print(f"{row['Player Name (English)']:30s} - PKR {row['Price (PKR)']:>15,} ({row['Price (Urdu)']})")

# Visualize top 10
plt.figure(figsize=(12, 6))
top_10_plot = df.nlargest(10, 'Price (PKR)').sort_values('Price (PKR)')
plt.barh(top_10_plot['Player Name (English)'], top_10_plot['Price (PKR)'])
plt.xlabel('Price (PKR)', fontsize=12)
plt.ylabel('Player Name', fontsize=12)
plt.title('Top 10 Most Expensive Players - HBL PSL 2026', fontsize=14, fontweight='bold')
plt.ticklabel_format(style='plain', axis='x')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))
plt.tight_layout()
plt.show()
```

## Price Distribution

```python
# Price distribution histogram
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['Price (PKR)'], bins=15, edgecolor='black', alpha=0.7, color='skyblue')
plt.xlabel('Price (PKR)', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.title('Price Distribution', fontsize=14, fontweight='bold')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))

plt.subplot(1, 2, 2)
plt.boxplot(df['Price (PKR)'], vert=True)
plt.ylabel('Price (PKR)', fontsize=12)
plt.title('Price Box Plot', fontsize=14, fontweight='bold')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.0f}M'))

plt.tight_layout()
plt.show()
```
