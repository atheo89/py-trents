# HBL PSL 2026 Player Prices Exploratory Data Analysis (EDA)
This notebook performs an EDA on the player prices for HBL PSL 2026.


## Import Libraries and Data


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Configure settings
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/datasets/babaralikhan098com/hbl-psl-auction-2026/list of cricket players and prices for HBL PSL 2026.csv')
df.head()
```

## Data Inspection and Cleaning
The dataset contains 27 players and 4 columns. The numeric price is stored as an object with commas. We will clean and convert it to a numeric type for analysis.


```python
df.info()
# Clean the 'Price (Numeric Estimate)' column
df['Price (Numeric Estimate)'] = df['Price (Numeric Estimate)'].str.replace(',', '').astype(float)
```

## Descriptive Statistics


```python
df.describe()
```

## Visualizations
### 1. Top 10 Most Expensive Players


```python
df_sorted = df.sort_values(by='Price (Numeric Estimate)', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_sorted.head(10), x='Price (Numeric Estimate)', y='Player Name (English)', palette='viridis')
plt.title('Top 10 Most Expensive Players in HBL PSL 2026')
plt.xlabel('Price (PKR)')
plt.ylabel('Player Name')
plt.show()
```

### 2. Distribution of Player Prices


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (Numeric Estimate)'], bins=10, kde=True, color='blue')
plt.title('Distribution of Player Prices in HBL PSL 2026')
plt.xlabel('Price (PKR)')
plt.ylabel('Frequency')
plt.show()
```

### 3. Cumulative Player Price Share
This shows how the total spending is concentrated among the players.


```python
df_sorted['Cumulative Price Share'] = df_sorted['Price (Numeric Estimate)'].cumsum() / df_sorted['Price (Numeric Estimate)'].sum()
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(df_sorted) + 1), df_sorted['Cumulative Price Share'], marker='o')
plt.title('Cumulative Player Price Share')
plt.xlabel('Number of Players')
plt.ylabel('Cumulative Price Percentage')
plt.grid(True)
plt.show()
```

## Conclusion
The analysis shows a clear concentration of high prices for top players, with a significant drop-off for others. Naseem Shah is currently the most expensive player on the list.
