```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/epstein-list-persons-of-interest/epstein-persons-2026-02-13_cleaned.csv")
```

```python
df.head()
```

```python
# Drop Ghislaine and Jeffrey
df = df[~df['Name'].isin(['Ghislaine Maxwell',  'Jeffrey Epstein'])]
```

```python
documents = df.sort_values("Documents", ascending=False)[:20]

sns.set_color_codes("pastel")
sns.barplot(x="Documents", y="Name", data=documents,
            label="Total", color="r",orient='h')

plt.title('Documents mentioning person in Epstein files', fontsize=16)
```

```python
flights = df.sort_values("Flights", ascending=False)[:20]

sns.set_color_codes("pastel")
sns.barplot(x="Flights", y="Name", data=flights,
            label="Total", color="b",orient='h')

plt.title('Flight on Epsteins Jet (1997-2006)', fontsize=16)
```

```python
connections = df.sort_values("Connections", ascending=False)[:20]

sns.set_color_codes("pastel")
sns.barplot(x="Connections", y="Name", data=connections,
            label="Total", color="y",orient='h')

plt.title('Number of Connections associated with Epstein', fontsize=16)
```

```python
nationality_counts = df['Nationality'].value_counts().reset_index()[:20]
nationality_counts.columns = ['Nationality', 'Count']  

sns.barplot(x='Count', y='Nationality', data=nationality_counts)

plt.title('Count of Nationalities', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Nationality', fontsize=12)
```

```python
category_counts = df['Category'].value_counts().reset_index()[:20]
category_counts.columns = ['Category', 'Count']  

sns.barplot(x='Count', y='Category', data=category_counts)

plt.title('Count of Category', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Nationality', fontsize=12)
```
