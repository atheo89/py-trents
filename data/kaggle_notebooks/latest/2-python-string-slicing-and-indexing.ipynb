### **Q1) Solution**

Here is the code and output for the word `'python'` based on the patterns established in your subsequent questions.

**Python Code:**

````
word = 'python'

# i) The character at position 0
print("Character in position 0 is " + word[0])

# ii) The character in position 5
print("Character in position 5 is " + word[5])

# iii) The last character
print("Last character is " + word[-1])

# iv) The second-to-last character
print("Second last character is " + word[-2])

# v) The characters from position 0 to 1
# (Note: Slice [0:2] includes index 0 and 1, excluding 2)
print("Characters from position 0 to 1 is " + word[0:2])
````

**Output:**

```text
Character in position 0 is p
Character in position 5 is n
Last character is n
Second last character is o
Characters from position 0 to 1 is py

```

### **Q3) Verification**

Below is the corrected syntax and output for splitting the string `"ComputingwithPython"`.

**Python Code:**

````
a = "ComputingwithPython"

print("First word is: " + a[0:9])
print("Second word is: " + a[9:13])
print("Third word is: " + a[13:19])

# Concatenating with spaces
print(a[0:9] + " " + a[9:13] + " " + a[13:19])
````

**Output:**

```text
First word is: Computing
Second word is: with
Third word is: Python
Computing with Python
```
