# ML Classification Models 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Loading, Exploration and Pre-processing

```python
import os
os.listdir("/kaggle/input/")
```

```python
# 1. Load the dataset
df = pd.read_csv("/kaggle/input/datasets/organizations/uciml/pima-indians-diabetes-database/diabetes.csv")
df
```

```python
# 2. Separate features and target
X = df.drop("Outcome", axis=1)
y = df["Outcome"]

# 3. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=22, stratify=y
)

# 4. Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled[:1], X_test_scaled[:1]
```

## Develop MLP ANN Classification Model

```python
# import MLPclassifier from sklearn
from sklearn.neural_network import MLPClassifier

# Build MLP ANN classification model 
mlp_model = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=3)

# Train ANN model
mlp_model.fit(X_train_scaled, y_train)

# Evaluate ANN model using test and train datasets
print("Training accuracy: ",mlp_model.score(X_train_scaled, y_train))
print("Test accuracy: ",mlp_model.score(X_test_scaled, y_test))
```

```python
# make a prediction using the trained classification model
# [Preg,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age]
new_data = [[6, 148, 72, 35, 0, 33.6, 0.627, 50]]  # example new data points
new_data_scaled = scaler.transform(new_data)
ANN_prediction = mlp_model.predict(new_data_scaled)
ANN_prediction
```

```python
# use widget to implement an UI to make prediction using the trained classification model
# get input from user using sliders for each feature
import ipywidgets as widgets
from IPython.display import display

# Feature names in order
feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Reasonable slider ranges for each feature (min, max, step)
slider_params = [
    (0, 20, 1),      # Pregnancies
    (0, 200, 1),     # Glucose
    (0, 140, 1),     # BloodPressure
    (0, 100, 1),     # SkinThickness
    (0, 900, 1),     # Insulin
    (10.0, 70.0, 0.1), # BMI
    (0.0, 2.5, 0.01),  # DiabetesPedigreeFunction
    (18, 100, 1)     # Age
]

sliders = [
    widgets.IntSlider(min=slider_params[0][0], max=slider_params[0][1], step=slider_params[0][2], description=feature_names[0]),
    widgets.IntSlider(min=slider_params[1][0], max=slider_params[1][1], step=slider_params[1][2], description=feature_names[1]),
    widgets.IntSlider(min=slider_params[2][0], max=slider_params[2][1], step=slider_params[2][2], description=feature_names[2]),
    widgets.IntSlider(min=slider_params[3][0], max=slider_params[3][1], step=slider_params[3][2], description=feature_names[3]),
    widgets.IntSlider(min=slider_params[4][0], max=slider_params[4][1], step=slider_params[4][2], description=feature_names[4]),
    widgets.FloatSlider(min=slider_params[5][0], max=slider_params[5][1], step=slider_params[5][2], description=feature_names[5]),
    widgets.FloatSlider(min=slider_params[6][0], max=slider_params[6][1], step=slider_params[6][2], description=feature_names[6]),
    widgets.IntSlider(min=slider_params[7][0], max=slider_params[7][1], step=slider_params[7][2], description=feature_names[7])
]

input_form = widgets.VBox(sliders)
display(input_form)

# Button to submit
submit_button = widgets.Button(description="Predict")
output = widgets.Output()

def on_submit_clicked(b):
    with output:
        output.clear_output()
        new_data = [[s.value for s in sliders]]
        new_data_scaled = scaler.transform(new_data)
        pred = mlp_model.predict(new_data_scaled)
        # For probability, use predict_proba if available
        if hasattr(mlp_model, 'predict_proba'):
            prob = mlp_model.predict_proba(new_data_scaled)[0][1]
            print(f"Predicted outcome (1=Diabetes, 0=No Diabetes): {pred[0]}")
            print(f"Probability of Diabetes: {prob:.2f}")
        else:
            print(f"Predicted outcome (1=Diabetes, 0=No Diabetes): {pred[0]}")
            print("Probability not available for this model.")

submit_button.on_click(on_submit_clicked)
display(submit_button, output)
```
