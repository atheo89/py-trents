```python
import yaml

data_config = {
    'path': '/kaggle/input/yolorant-yolo26',
    'train': 'train/images',
    'val': 'valid/images',
    'test': 'test/images',
    'nc': 2,
    'names': ['enemy', 'enemy_head']
}

with open('/kaggle/working/yolorant_data.yaml', 'w') as f:
    yaml.dump(data_config, f)

data_config
```

```python
from ultralytics import YOLO
import os

model = YOLO('yolo26n.pt')
results = model.train(
    data='/kaggle/working/yolorant_data.yaml', 
    epochs=100,
    imgsz=512,
    batch=-1,
    device=0,
    workers=4,
    name='y26n_512',

    optimizer='AdamW',    
    lr0=0.001,            
    augment=False,
    val=True,             
    amp=True,             
    patience=20,
    
    mosaic=0.0,
    mixup=0.0,
    close_mosaic=10
)
```
