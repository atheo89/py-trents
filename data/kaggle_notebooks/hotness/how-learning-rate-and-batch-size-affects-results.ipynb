## 1) What is Learning Rate? (Simple Meaning)

The **learning rate** is the size of each step your model takes while learning.

Think of training like walking down a mountain to reach the lowest point (**minimum loss**).

* **High learning rate** = huge jumps 🦘
* **Low learning rate** = tiny steps 🐢

---

## 2) Case 1 — Learning Rate Too High 

###  What you observe

When learning rate is too high:

* The **loss curve bounces up and down**
* Loss does **not steadily decrease**
* Model predictions look **bad**
* Training feels unstable

###  Why it happens

Because the model takes **too large steps**, it overshoots the best solution again and again.

###  Beginner Example

Imagine you want to reach a door, but every step you take is 10 meters.

You will:

* jump past the door
* jump back
* jump again
* never land exactly near it

---

## 3) Case 2 — Learning Rate Too Low 

###  What you observe

When learning rate is too small:

* Loss decreases **very slowly**
* It may not converge in 20 epochs
* Model improves, but takes too long

###  Why it happens

Because the model takes **tiny steps**, it moves towards the best solution but slowly.

###  Beginner Example

You want to reach a door 10 meters away.

But each step is only 1 cm.

You will reach it eventually… but it takes a lot of time.

---

## 4) Case 3 — Batch Size Effect 

### What is Batch Size?

Batch size means:

> How many training examples are used before the model updates its weights.

Example:

* Batch size = 1 → update after every data point
* Batch size = 32 → update after every 32 samples
* Batch size = 128 → update after every 128 samples

---

## 5) What Happens When Batch Size is Increased?

###  What you observe

When batch size is increased:

* Each epoch runs **faster** ⏩
* But the model may not converge quickly in few epochs


```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

np.random.seed(42)

X = np.random.rand(200, 1)
Y = 3 * X + 2 + 0.2 * np.random.randn(200, 1)

def build_model(lr=0.01):
    model = keras.Sequential([
        keras.layers.Dense(1, input_shape=(1,))
    ])

    model.compile(
        optimizer=keras.optimizers.SGD(learning_rate=lr),
        loss='mse'
    )
    return model

learning_rates = [1.0, 0.1, 0.01, 0.001]

histories = {}

for lr in learning_rates:
    print(f"Training with learning rate = {lr}")
    model = build_model(lr=lr)
    history = model.fit(X, Y, epochs=20, batch_size=32, verbose=0)
    histories[lr] = history.history['loss']

plt.figure(figsize=(10, 6))

for lr, loss in histories.items():
    plt.plot(loss, label=f"lr={lr}")

plt.title("Loss Curves for Different Learning Rates")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

batch_sizes = [8, 32, 128]

histories_bs = {}

for bs in batch_sizes:
    print(f"Training with batch size = {bs}")
    model = build_model(lr=0.01)
    history = model.fit(X, Y, epochs=20, batch_size=bs, verbose=0)
    histories_bs[bs] = history.history['loss']

plt.figure(figsize=(10, 6))

for bs, loss in histories_bs.items():
    plt.plot(loss, label=f"batch_size={bs}")

plt.title("Loss Curves for Different Batch Sizes")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```
