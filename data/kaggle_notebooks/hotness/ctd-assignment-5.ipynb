## Task 1

```python
df1=pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 30],
    'Salary': [50000, 60000, 70000, 80000, 55000]
})

df2=pd.DataFrame({
    'Name': ['Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Age': [28, 33, 35, 29, 40],
    'Salary': [52000, 58000, 72000, 61000, 85000]
})

df3=pd.DataFrame({
    'Name': ['Frank', 'Helen', 'Ian', 'Hima', 'Chaka'],
    'Age': [17, 93, 12, 57, 106],
    'Favorite Color': ['blue', 'pink', 'burgundy', 'red', 'turquoise']
})
print(df1,'\n',df2,'\n',df3)
```

```python
names=df1["Name"]
print(names)

nameandsalary=df1[['Name', 'Salary']]
print(nameandsalary)

print(df1.iloc[:3])
```

## Task 2

```python
groupedmean = df1.groupby('Age')['Salary'].mean()
print(groupedmean) 

groupedsum = df1.groupby('Age')['Salary'].sum()
print(groupedsum) 

groupedcount = df1.groupby('Age')['Salary'].count()
print(groupedcount) 
```

## Task 3


```python
with np.errstate(invalid='ignore'):
    
    df_1_3_merged=pd.merge(df1,df3,on=['Name'],how='outer',suffixes=['_left','_right'])
    print(df_1_3_merged)
    df_1_3_merged['Salary'] = df_1_3_merged['Salary'].fillna(15000)
    df_1_3_merged['Favorite Color'] = df_1_3_merged['Favorite Color'].fillna('yellow')
    print(df_1_3_merged)
    df_1_3_merged['Age'] = np.where(df_1_3_merged['Age_left'].notna(),df_1_3_merged['Age_left'], df_1_3_merged['Age_right'])
    print(df_1_3_merged)
    df_1_3_merged.drop(['Age_left','Age_right'], axis=1, inplace=True)
    print(df_1_3_merged)
    df1_b = df1.copy()
    df1_b.set_index('Name', inplace=True)
    df3_b = df3.copy()
    df3_b.set_index('Name', inplace=True)
    joined_df = df1_b.join(df3_b, how='outer',lsuffix='_left', rsuffix='_right')
    print(joined_df)
```

## Task 9

```python
football_results=pd.read_csv("/kaggle/input/international-football-results-from-1872-to-2017/results.csv")
print(football_results.head())
```

```python
results_1=football_results[["home_team","away_team","home_score","away_score","date"]]
print(results_1.head())
```

```python
results_2=results_1.copy()
results_2=results_2.rename(columns={'home_team':'team','away_team':'opponent'})
print(results_2.head())
```

```python
results_3=results_1.copy()
results_3=results_3.rename(columns={'home_team':'opponent','away_team':'team','away_score':'points_for','home_score':'points_against'})
print(results_3.head())
```

```python
football_results=pd.concat([results_2, results_3], ignore_index=True)
print(football_results.head())
```

```python

points_against = football_results.groupby('team')['points_against'].mean()

# 2. Sort and SAVE the result back to the variable
points_against = points_against.sort_values(ascending=False)

# 3. Print top 10
print(points_against.head(10))
```

```python
tunisia=football_results[football_results['team'] == "Tunisia"]
tunisia_recent=tunisia.sort_values(by='date',ascending=False)
print(tunisia_recent.head(10))
```
