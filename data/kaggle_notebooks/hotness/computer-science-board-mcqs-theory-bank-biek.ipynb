# 5. Exploratory Data Analysis (EDA)

## 1. Analysis by Year 

```python
year_counts = df['year'].value_counts().sort_index()

plt.figure()
plt.bar(year_counts.index, year_counts.values,color='orange')
plt.title("Questions Distribution by Exam Year")
plt.xlabel("Exam Year")
plt.ylabel("Number of Questions")
plt.show()
```

## 2. Analysis by Type

```python
type_counts = df['type'].value_counts()

plt.figure()
plt.bar(type_counts.index, type_counts.values, width=0.2,color='red')
plt.title("Distribution of Types")
plt.xticks(rotation=90)
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.plot(year_counts.index, year_counts.values)
plt.title("Year vs Question Count (Scatter)")
plt.xlabel("Exam Year")
plt.ylabel("Number of Questions")
plt.show()
```

## 3 Question Of Type in Percentage

```python
plt.figure()
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%')
plt.title("Question Type Percentage")
plt.show()
```
