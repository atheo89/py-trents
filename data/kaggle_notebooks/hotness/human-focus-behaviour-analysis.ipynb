```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("/kaggle/input/background-noise-and-focus-dataset/background_noise_focus_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
fig = px.histogram(
    df,
    x="perceived_focus_score",
    title="Distribution of Perceived Focus Score"
)
fig.show()
```

```python
fig = px.box(
    df,
    x="role",
    y="age",
    color="role",
    title="Age Distribution by Role"
)
fig.show()
```

```python
fig = px.bar(
    df,
    x="background_noise_type",
    y="perceived_focus_score",
    color="background_noise_type",
    title="Focus Score by Noise Type",
)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="noise_volume_level",
    y="focus_duration_minutes",
    color="background_noise_type",
    size="task_completion_quality",
    title="Noise Volume vs Focus Duration"
)
fig.show()
```

```python
fig = px.violin(
    df,
    x="task_type",
    y="mental_fatigue_after_task",
    color="task_type",
    box=True,
    title="Mental Fatigue Distribution by Task"
)
fig.show()
```

```python
plt.figure()
plt.scatter(df["focus_duration_minutes"], df["perceived_focus_score"])
plt.xlabel("Focus Duration")
plt.ylabel("Focus Score")
plt.title("Focus Duration vs Perceived Focus Score")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
heatmap_data = pd.crosstab(df["role"], df["task_type"])

fig = px.imshow(
    heatmap_data,
    title="Role vs Task Type Heatmap"
)
fig.show()
```

```python
fig = px.box(
    df,
    x="background_noise_type",
    y="task_completion_quality",
    color="background_noise_type",
    title="Task Completion Quality by Noise Type"
)
fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="noise_volume_level",
    y="focus_duration_minutes",
    z="perceived_focus_score",
    color="background_noise_type",
    title="3D Focus Behaviour Visualization"
)
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Histogram(x=df["focus_duration_minutes"], name="Focus Duration"))
fig.add_trace(go.Histogram(x=df["noise_volume_level"], name="Noise Volume"))

fig.update_layout(title="Focus Duration vs Noise Volume Distribution")
fig.show()
```

## Conclusion:

The visualization analysis shows that background noise type and noise volume have a noticeable influence on focus duration and perceived focus levels. Tasks performed in silence or low-volume environments generally demonstrate better focus consistency compared to traffic noise or lyrical music. Mental fatigue tends to increase when focus duration is longer or task complexity is higher. Role and task type also create variations in performance patterns, indicating behavioral differences among participants. Overall, the dataset suggests that environmental conditions and task characteristics jointly impact productivity and concentration outcomes.
