## 1 Import Libraries <a name="1 Import Libraries"></a>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2 Loading the Datase <a name="2 Loading the Datase"></a>

```python
df = pd.read_csv('/kaggle/input/loan-approval-prediction-dataset/loan_approval_dataset.csv')
```

```python
df.columns = df.columns.str.strip()
```

## 4 Remove Unimportant Columns <a name="4 Remove Unimportant Columns"></a>

```python
df = df.drop(columns = "loan_id")
```

## 5 Encoding <a name="5 Encoding"></a>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == "object":
        df[col] = le.fit_transform(df[col])
```

## 6 Split Data <a name="6 Split Data"></a>

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns = "loan_status")
y = df["loan_status"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# 7 Treatment of imbalances using SMOTE <a name="7 Treatment of imbalances using SMOTE"></a>

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

# 8 Modelling Decision Tree <a name="8 Modelling Decision Tree"></a>

```python
from sklearn.tree import DecisionTreeClassifier
DT  = DecisionTreeClassifier(max_depth=10, random_state=42)
DT.fit(x_train_sm, y_train_sm)
target = DT.predict(x_test)
```

# 9 Evaluation Decision Tree <a name="9 Evaluation Decision Tree"></a>

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test, target))
```

```python
print(confusion_matrix(y_test, target))
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision, recall, f1, _ = precision_recall_fscore_support(y_test, target, average=None)
Classes = ["Approved", "Rejected"]
metrics = pd.DataFrame({
    "Class": Classes,
    "Precision": precision,
    "Recall": recall,
    "F1-score": f1
})
metrics
```

# 10 Modelling Logistic Regression <a name="10 Modelling LogisticRegression"></a>

```python
from sklearn.linear_model import LogisticRegression
lo = LogisticRegression(max_iter = 1500)
lo.fit(x_train_sm, y_train_sm)
target = lo.predict(x_test)
```

# 11 Evaluation Logistic Regression <a name="11 Evaluation Logistic Regression"></a>

```python
print(accuracy_score(y_test, target))
```

```python
print(confusion_matrix(y_test, target))
```

```python
precision, recall, f1, _ = precision_recall_fscore_support(y_test, target, average=None)
Classes = ["Approved", "Rejected"]
metrics = pd.DataFrame({
    "Class": Classes,
    "Precision": precision,
    "Recall": recall,
    "F1-score": f1
})
metrics
```
