# Define Global Configuration

```python
# ==========================================
# STEP 7: GLOBAL CONFIGURATION
# ==========================================

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

train_dir = "/kaggle/working/Merged_Dataset/Training"
test_dir = "/kaggle/working/Merged_Dataset/Testing"

print("Configuration set successfully.")
```

# Import tensorflow

```python
# ==========================================
# IMPORT TENSORFLOW & KERAS
# ==========================================

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow version:", tf.__version__)
```

```python
print("GPU Available:", tf.config.list_physical_devices('GPU'))
```

```python
# ==========================================
# STEP 8: LOAD DATASETS FROM DIRECTORY
# ==========================================

train_ds = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    labels="inferred",
    label_mode="int",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=True,
    seed=SEED
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    labels="inferred",
    label_mode="int",
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

print("Datasets loaded successfully.")
```

```python
class_names = train_ds.class_names
print("Class names:", class_names)
```

```python
for images, labels in train_ds.take(1):
    print(images.shape)
```

# NORMALIZATION + PERFORMANCE OPTIMIZATION

```python
# ==========================================
# STEP 9: NORMALIZATION + PERFORMANCE OPTIMIZATION
# ==========================================

AUTOTUNE = tf.data.AUTOTUNE

# Normalize pixel values (0-255 → 0-1)
normalization_layer = layers.Rescaling(1./255)

train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y),
                        num_parallel_calls=AUTOTUNE)

test_ds = test_ds.map(lambda x, y: (normalization_layer(x), y),
                      num_parallel_calls=AUTOTUNE)

# Cache and prefetch for performance
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)

print("Data pipeline optimized successfully.")
```

# Define Augmentation Layer

```python
# ==========================================
# STEP 10: DATA AUGMENTATION
# ==========================================

data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
])

print("Data augmentation layer created.")
```

# CNN ARCHITECTURE DESIGN

```python
# ==========================================
# STEP 11: CNN MODEL ARCHITECTURE
# ==========================================

model = keras.Sequential([

    # Data Augmentation
    data_augmentation,
    
    # First Conv Block
    layers.Conv2D(32, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    
    # Second Conv Block
    layers.Conv2D(64, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    
    # Third Conv Block
    layers.Conv2D(128, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    
    # Fourth Conv Block
    layers.Conv2D(256, (3, 3), padding="same", activation="relu"),
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    
    # Flatten
    layers.GlobalAveragePooling2D(),
    
    # Dense Layer
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    
    # Output Layer (4 classes)
    layers.Dense(4, activation="softmax")
])

print("CNN model created successfully.")
```

# create validation split

```python
# ==========================================
# STEP 12: CREATE VALIDATION SPLIT FROM train_ds
# ==========================================

import tensorflow as tf

# Calculate number of batches
train_batches = tf.data.experimental.cardinality(train_ds).numpy()
print("Total training batches:", train_batches)

# Define validation fraction (20%)
val_fraction = 0.20
val_batches = max(1, int(val_fraction * train_batches))
train_batches_final = train_batches - val_batches

print(f"Using {val_batches} batches for validation and {train_batches_final} for training.")

# Create datasets
val_ds = train_ds.take(val_batches)
train_ds_final = train_ds.skip(val_batches)

# Prefetch already applied, but ensure AUTOTUNE is set
AUTOTUNE = tf.data.AUTOTUNE
train_ds_final = train_ds_final.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

print("Validation split created.")
```

# Compile the Model + Callbacks (EarlyStopping, ReduceLROnPlateau, Checkpoint)

```python
# ==========================================
# STEP 13: COMPILE MODEL + CALLBACKS
# ==========================================

from tensorflow.keras import callbacks

# Hyperparameters
initial_lr = 1e-4
epochs = 30

# Compile
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=initial_lr),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
es = callbacks.EarlyStopping(
    monitor='val_loss',
    patience=6,
    restore_best_weights=True,
    verbose=1
)

rlrop = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=3,
    min_lr=1e-7,
    verbose=1
)

ckpt_path = "/kaggle/working/best_cnn_model.h5"
mc = callbacks.ModelCheckpoint(
    ckpt_path,
    monitor='val_loss',
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)

print("Model compiled with Adam lr=", initial_lr)
print("Callbacks: EarlyStopping, ReduceLROnPlateau, ModelCheckpoint")
```

# Train the Model (Fit)

```python
# ==========================================
# STEP 14: MODEL TRAINING
# ==========================================

history = model.fit(
    train_ds_final,
    validation_data=val_ds,
    epochs=epochs,
    callbacks=[es, rlrop, mc],
    verbose=1
)
```

```python
# ==========================================
# STEP 15: EVALUATE ON TEST DATA
# ==========================================

test_loss, test_accuracy = model.evaluate(test_ds)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# ==========================================
# STEP 16: CONFUSION MATRIX
# ==========================================

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Get predictions
y_true = []
y_pred = []

for images, labels in test_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

# Convert to numpy
y_true = np.array(y_true)
y_pred = np.array(y_pred)

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Classification report
print(classification_report(y_true, y_pred, target_names=class_names))
```
