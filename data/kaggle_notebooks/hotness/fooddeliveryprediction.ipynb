```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
```

```python
data=pd.read_csv("/kaggle/input/food-delivery-time-prediction/Food_Delivery_Times.csv")
```

```python
data=data.drop("Order_ID", axis=1)
```

```python
num_col=["Distance_km","Preparation_Time_min","Courier_Experience_yrs","Delivery_Time_min"]
cat_col=["Weather","Traffic_Level","Time_of_Day","Vehicle_Type"]
```

```python
for col in num_col:
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
for col in cat_col:
    data[col].fillna(data[col].mode(), inplace=True)
```

```python
le=LabelEncoder()
for col in cat_col:
    data[col]=le.fit_transform(data[col])
```

```python
data.head
```

```python
x=data.drop("Delivery_Time_min", axis=1)
Y=data["Delivery_Time_min"]
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(x)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.3)
```

```python
model=KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, Y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean: ",mean_absolute_error(Y_test, y_pred))
```

```python
plt.figure()
plt.scatter(Y_test, y_pred)
plt.xlabel("Actual Delivery Time")
plt.ylabel("Predicted Delivery Time")
plt.title("Delivery Time")
plt.show()
```
