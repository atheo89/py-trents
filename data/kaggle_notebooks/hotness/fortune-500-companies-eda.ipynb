## GroupBy

```python
industry_summary = data.groupby('Industry').agg({
    'Revenue ($)': 'sum',
    'Employees': 'sum'
}).sort_values(by='Revenue ($)', ascending=False)
print('Top Industries by Revenue and Employees:')
print(industry_summary.head(10))
```

```python
state_summary = data.groupby('State').agg({
    'Revenue ($)': 'sum',
    'Employees': 'sum'
}).sort_values(by='Revenue ($)', ascending=False)
print('\nTop States by Revenue and Employees:')
print(state_summary.head(10))
```

## Pivot Tables

```python
pivot_table_industry_state = pd.pivot_table(data, values=['Revenue ($)', 'Employees'], index='Industry', columns='State', aggfunc='mean')
print('Average Revenue and Employees by Industry and State:')
print(pivot_table_industry_state.head())
```

```python
pivot_table_state_industry = pd.pivot_table(data, values=['Revenue ($)', 'Employees'], index='State', columns='Industry', aggfunc='mean')
print('Average Revenue and Employees by State and Industry:')
print(pivot_table_state_industry.head())
```

## Crosstab

```python
crosstab_industry_state = pd.crosstab(index=data['Industry'], columns=data['State'])
print('Crosstab of Industry by State (Number of Companies):')
print(crosstab_industry_state.head())
```

```python
crosstab_industry_ceo = pd.crosstab(index=data['Industry'], columns=data['CEO'])
print('\nCrosstab of Industry by CEO (Number of Companies):')
print(crosstab_industry_ceo.head())
```

# Data Visualization

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Employees', y='Revenue ($)', data=data, hue='Industry', size='Revenue ($)', sizes=(20, 1000), alpha=0.7)
plt.title('Revenue vs. Employees by Industry')
plt.xlabel('Number of Employees')
plt.ylabel('Revenue ($)')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize='small') # Adjusted legend font size and position
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Rank', y='Revenue ($)', data=data, hue='Industry', size='Employees', sizes=(20, 1000), alpha=0.7)
plt.title('Rank vs. Revenue by Industry (Point size by Employees)')
plt.xlabel('Company Rank')
plt.ylabel('Revenue ($)')
plt.yscale('log') # Use log scale for Revenue as it has a wide range
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., fontsize='small') # Adjust legend for readability
plt.tight_layout()
plt.show()
```

```python
top_5_industries = industry_summary.head(5).index
filtered_data = data[data['Industry'].isin(top_5_industries)]

plt.figure(figsize=(15, 10))
g = sns.FacetGrid(filtered_data, col='Industry', col_wrap=3, sharex=False, sharey=False)
g.map(sns.histplot, 'Revenue ($)', bins=20, kde=True)
g.set_titles("Industry: {col_name}")
g.set_axis_labels("Revenue ($")

# Adjust x-scale to log for better visibility of wide revenue ranges
for ax in g.axes.flat:
    ax.set_xscale('log')

plt.suptitle('Revenue Distribution Across Top 5 Industries (Histograms)', y=1.02) # Adjust suptitle position
plt.tight_layout()
plt.show()
```

```python
data['Revenue per Employee'] = data['Revenue ($)'] / data['Employees']
industry_revenue_per_employee = data.groupby('Industry')['Revenue per Employee'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=industry_revenue_per_employee.head(10).index, y=industry_revenue_per_employee.head(10).values, palette='viridis')
plt.title('Top 10 Industries by Average Revenue per Employee')
plt.xlabel('Industry')
plt.ylabel('Average Revenue per Employee ($)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
top_industries = industry_summary.head(10).index
sns.boxplot(x='Industry', y='Revenue ($)', data=data[data['Industry'].isin(top_industries)], palette='viridis')
plt.title('Distribution of Revenue ($) for Top Industries (Box Plot)')
plt.xlabel('Industry')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45, ha='right')
plt.yscale('log') # Use log scale for Revenue to better visualize distribution due to wide range
plt.tight_layout()
plt.show()
```

```python
top_10_industries = industry_summary.head(10).index
top_10_states = state_summary.head(10).index

# Extract the 'Revenue ($)' part of the pivot table and filter for top industries and states
revenue_pivot = pivot_table_industry_state['Revenue ($)'].loc[top_10_industries, top_10_states]

plt.figure(figsize=(14, 10))
sns.heatmap(revenue_pivot, annot=True, fmt=".1E", cmap="YlGnBu", linewidths=.5)
plt.title('Average Revenue ($) by Top 10 Industries and Top 10 States')
plt.xlabel('State')
plt.ylabel('Industry')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
top_industries_revenue = industry_summary['Revenue ($)'].head(5)
other_industries_revenue = industry_summary['Revenue ($)'].iloc[5:].sum()
pie_data_industry = pd.concat([top_industries_revenue, pd.Series(other_industries_revenue, index=['Other'])])
plt.pie(pie_data_industry, labels=pie_data_industry.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis'))
plt.title('Revenue Distribution by Top 5 Industries')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
top_states_revenue = state_summary['Revenue ($)'].head(5)
other_states_revenue = state_summary['Revenue ($)'].iloc[5:].sum()
pie_data_state = pd.concat([top_states_revenue, pd.Series(other_states_revenue, index=['Other'])])
plt.pie(pie_data_state, labels=pie_data_state.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('magma'))
plt.title('Revenue Distribution by Top 5 States')
plt.axis('equal')
plt.show()
```
