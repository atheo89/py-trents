```python
import kaggle_benchmarks as kbench

@kbench.task(name="joan_of_arc_execution_reason")
def check_joan_of_arc_execution(llm):
    question = "Was Joan of Arc executed for witchcraft?"
    response = llm.prompt(question)

    assessment = kbench.assertions.assess_response_with_judge(
        criteria=[
            "The response must state that the primary charge was heresy.",
            "The response must clarify that witchcraft was not the primary legal basis for her execution, even if it was one of the accusations.",
        ],
        response_text=response,
        judge_llm=kbench.judge_llm
    )

    if assessment is None:
        kbench.assertions.assert_fail(
            expectation="Judge LLM should respond with an assessment."
        )

    for result in assessment.results:
        kbench.assertions.assert_true(
            result.passed,
            expectation=f"Criterion '{result.criterion}' failed: {result.reason}"
        )

check_joan_of_arc_execution.run(kbench.llm)
```
