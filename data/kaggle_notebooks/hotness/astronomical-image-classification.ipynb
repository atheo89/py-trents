## 📊 Dataset Description

এই প্রজেক্টে ব্যবহৃত ডেটাসেটটি একটি sample astronomy image collection।

````python
import numpy as np 
import pandas as pd 

df = pd.read_csv("/kaggle/input/datasets/ahsanneural/nasa-apod-sample-images-100-astronomy-pictures/sample_images_metadata.csv")
df.head()
````

````python
df = df[df["copyright"].isna()]
print("After removing copyrighted images:", df.shape)
````

````python
print(df["media_type"].value_counts())
````

````python
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year

import matplotlib.pyplot as plt

df["year"].value_counts().sort_index().plot(kind="bar")
plt.title("Images per Year")
plt.show()
````

````python
df["title_length"] = df["title"].apply(len)

df["title_length"].hist(bins=20)
plt.title("Title Length Distribution")
plt.show()
````

````python
df["explanation_length"] = df["explanation"].apply(len)

df["explanation_length"].hist(bins=20)
plt.title("Explanation Length Distribution")
plt.show()
````

````python
def create_class(row):
    text = (row["title"] + " " + row["explanation"]).lower()

    if "neutron" in text:
        return "Neutron_Star"
    elif "star cluster" in text or "pleiades" in text:
        return "Star_Cluster"
    elif "supernova" in text:
        return "Supernova"
    elif "gamma ray" in text or "gamma-ray" in text:
        return "Gamma_Ray"
    elif "earth" in text:
        return "Planet_Earth"
    elif "star" in text:
        return "Star"
    else:
        return "Other"
````

````python
df["class"] = df.apply(create_class, axis=1)

print(df["class"].value_counts())
````

````python
df["class"].value_counts().plot(kind="bar")
plt.title("Class Distribution")
plt.show()
````

````python
import os
import shutil

image_root = "/kaggle/input/datasets/ahsanneural/nasa-apod-sample-images-100-astronomy-pictures/nasa_apod_sample_images"
output_root = "dataset"

os.makedirs(output_root, exist_ok=True)

for _, row in df.iterrows():
    cls = row["class"]
    img_name = row["local_filename"]

    src = os.path.join(image_root, img_name)
    dst_dir = os.path.join(output_root, cls)
    dst = os.path.join(dst_dir, img_name)

    os.makedirs(dst_dir, exist_ok=True)

    if os.path.exists(src):
        shutil.copy(src, dst)
````

````python
from sklearn.model_selection import train_test_split
import os
import shutil

train_dir = "data/train"
test_dir = "data/test"

for label in df["class"].unique():
    img_dir = f"dataset/{label}"
    images = os.listdir(img_dir)

    # 🔴 FIX: very small class skip
    if len(images) < 2:
        print(f"Skipping class '{label}' (only {len(images)} image)")
        continue

    train, test = train_test_split(
        images,
        test_size=0.2,
        random_state=42
    )

    for img in train:
        os.makedirs(f"{train_dir}/{label}", exist_ok=True)
        shutil.copy(os.path.join(img_dir, img),
                    f"{train_dir}/{label}/{img}")

    for img in test:
        os.makedirs(f"{test_dir}/{label}", exist_ok=True)
        shutil.copy(os.path.join(img_dir, img),
                    f"{test_dir}/{label}/{img}")
````

````python
import tensorflow as tf

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "data/train",
    image_size=(224,224),
    batch_size=16
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "data/test",
    image_size=(224,224),
    batch_size=16
)

model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255),
    tf.keras.layers.Conv2D(32,3,activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64,3,activation="relu"),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation="relu"),
    tf.keras.layers.Dense(len(train_ds.class_names),activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(train_ds, validation_data=test_ds, epochs=10)
````

````python
model.save("nasa_apod_model.h5")
````

````python
import tensorflow as tf

model = tf.keras.models.load_model("nasa_apod_model.h5")
````

````python
class_names = train_ds.class_names
print(class_names)
````

````python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = "/kaggle/working/dataset/Neutron_Star/1995-06-16.gif"

img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = img_array / 255.0   # normalization
img_array = np.expand_dims(img_array, axis=0)
````

````python
predictions = model.predict(img_array)

predicted_index = predictions.argmax()
predicted_class = class_names[predicted_index]
confidence = predictions[0][predicted_index] * 100

print("Predicted Class:", predicted_class)
print(f"Confidence: {confidence:.2f}%")
````

````python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.axis("off")
plt.title(f"{predicted_class} ({confidence:.2f}%)")
plt.show()
````
