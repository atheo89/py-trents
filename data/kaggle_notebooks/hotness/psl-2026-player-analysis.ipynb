The **Pakistan Super League (PSL) 2026** (Season 11) marks a significant milestone in the evolution of T20 cricket in Pakistan. As the league grows, so does the financial scale of player acquisitions. This notebook provides a comprehensive exploratory data analysis (EDA) of the player prices for the upcoming season, focusing on the elite bracket of players and their respective market values.

**Objectives:**
1. Analyze the distribution of player prices.
2. Identify the highest-earning players in the 2026 draft/retention.
3. Provide interactive visualizations to understand the financial hierarchy of the league.

---

## PSL ANALYSIS ABOUT PLAYERS

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings

warnings.filterwarnings('ignore')

# Set Plotly template for professional aesthetic
template = "plotly_dark"
main_color = "#00CC96"

print("Setup complete! Ready to transform data into insights.")
```

### 📁 2. Data Loading & Inspection
We load the dataset containing player names in English and Urdu, along with their estimated numeric prices in PKR.

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/datasets/muhammadaammartufail/hbl-psl-2026-cricket-players-auction-prices/list of cricket players and prices for HBL PSL 2026.csv')

# Display basic info and first few rows
print(f"Dataset Shape: {df.shape}")
df.head()
```

### 🧹 3. Data Cleaning & Preprocessing
The numeric prices are currently stored as strings with commas. We need to convert them to actual integers for mathematical analysis.

```python
# Rename columns for easier access
df.columns = ['Name_Urdu', 'Name', 'Price_Urdu', 'Price_PKR']

# Clean Price_PKR: Remove commas and convert to numeric
df['Price_PKR'] = df['Price_PKR'].str.replace(',', '').astype(float)

# Add a column for Price in Millions for better readability in charts
df['Price_Millions'] = df['Price_PKR'] / 1_000_000

df = df.sort_values(by='Price_PKR', ascending=False).reset_index(drop=True)

print("Data cleaning successful! Numeric values normalized.")
df.head()
```

### 📊 4. Exploratory Data Analysis (EDA)

#### 4.1 Statistical Overview
Let's look at the basic statistics of player prices.

```python
stats = df['Price_PKR'].describe()
print("--- Salary Statistics (PKR) ---")
print(stats)
```

#### 4.2 Top 10 Highest Valued Players
Who are the biggest earners in PSL 2026? Let's visualize the elite group.

```python
top_10 = df.head(10)

fig1 = px.bar(
    top_10, 
    x='Name', 
    y='Price_Millions', 
    text='Price_Millions', 
    title='🏆 Top 10 Highest Valued Players in PSL 2026',
    labels={'Price_Millions': 'Price (Millions PKR)', 'Name': 'Player Name'},
    template=template,
    color='Price_Millions',
    color_continuous_scale='Viridis'
)

fig1.update_traces(texttemplate='%{text:.1f}M', textposition='outside')
fig1.update_layout(
    xaxis_tickangle=-45,
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)',
    font=dict(family="Arial", size=12)
)

fig1.show()
```

#### 4.3 Distribution of Player Prices
How is the budget spread across the players? This histogram shows the concentration of player valuations.

```python
fig2 = px.histogram(
    df, 
    x='Price_Millions', 
    nbins=10,
    title='💹 Distribution of Player Prices',
    labels={'Price_Millions': 'Price Range (Millions PKR)'},
    template=template,
    marginal="box",  # Adds a box plot above the histogram
    color_discrete_sequence=[main_color]
)

fig2.update_layout(
    bargap=0.1,
    plot_bgcolor='rgba(0,0,0,0)',
    paper_bgcolor='rgba(0,0,0,0)'
)

fig2.show()
```

#### 4.4 Cumulative Value Contribution
This chart illustrates the "Head vs Tail" distribution of the league's player costs.

```python
df['Cumulative_Sum'] = df['Price_Millions'].cumsum()
total_value = df['Price_Millions'].sum()
df['Percentage_Contribution'] = (df['Cumulative_Sum'] / total_value) * 100

fig3 = go.Figure()

# Add Bar Chart for Individual Prices
fig3.add_trace(go.Bar(
    x=df['Name'],
    y=df['Price_Millions'],
    name='Individual Price',
    marker_color='#3498db'
))

# Add Line for Cumulative Percentage
fig3.add_trace(go.Scatter(
    x=df['Name'],
    y=df['Percentage_Contribution'],
    name='Cumulative %',
    yaxis='y2',
    line=dict(color='#e74c3c', width=3)
))

fig3.update_layout(
    title='📈 Pareto Analysis of Player Valuations',
    xaxis=dict(title='Players (Ranked Highest to Lowest)'),
    yaxis=dict(title='Price (Millions PKR)'),
    yaxis2=dict(title='Cumulative % of Total Budget', overlaying='y', side='right', range=[0, 105]),
    template=template,
    xaxis_tickangle=-90,
    legend=dict(x=0.01, y=0.99)
)

fig3.show()
```

### 📝 5. Key Insights & Strategy Recommendations

Based on the analysis of the PSL 2026 player data:

1. **Concentration of Wealth:** The top 5 players (Naseem Shah, Faheem Ashraf, Daryl Mitchell, Fakhar Zaman, and David Warner) represent a significant portion of the total player expenditure among the elite list. These players command salaries exceeding 79M PKR.
2. **Market Extremes:** We see a massive range in valuations, from a high of **86.5M PKR (Naseem Shah)** to a floor of **6M PKR (Jahanzeb Sultan)** within this specific dataset.
3. **Utility Clusters:** A clear "Middle Class" of players exists in the 20M - 50M PKR range, providing the essential balance for franchise rosters.
4. **Franchise Strategy:** The data suggests a "Stars and Scrubs" strategy where franchises invest heavily in pace-bowling spearheads (Naseem) and versatile all-rounders (Faheem, Daryl Mitchell).
