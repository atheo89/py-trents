```python
import numpy as np
import tensorflow
from tensorflow import keras
```

```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token = '<nothing>')

# By setting oov_token = '<nothing>', you are telling the Tokenizer: "If you encounter a word you don't recognize, don't crash and don't ignore it. 
# Instead, replace it with the special token <nothing>."
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index

# it gives every unique word (string) found in the text you provided to .fit_on_texts().
# Also, provides a unique integer (ID) assigned to that word.
```

```python
# we can also check that how many times a word is repeating :- 

tokenizer.word_counts
```

```python
# for counting the 'no. of sentences' or 'no. of rows present' in our document :-  

tokenizer.document_count
```

```python
# now, we have to generate the sequence for each sentence :- 

sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
# so here we can see that our sequence is generated i.e., [10, 2] -> shows our 1st sentence "go india"  and so on....
```

```python
# since here size of sequences are different, so first we have to make their sizes equal 
# that's why we have to use "Padding" to add zeros in the end (to make the length of each sequence similar).  
```

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')   

# 'post' -> it shows that zeros are adding at the last of the sequence.
```

```python
# print our sequences :-

sequences
```
