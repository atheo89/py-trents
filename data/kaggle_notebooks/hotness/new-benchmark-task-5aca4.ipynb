```python
import kaggle_benchmarks as kbench

@kbench.task(name="evaluate_model_improvement")
def evaluate_model_improvement(llm):
    prompt = """
Evaluate whether a machine learning model genuinely outperforms a baseline benchmark in a Kaggle competition.

Given:
- Baseline Cross-Validation Mean Score = 0.812
- Baseline Standard Deviation = 0.004
- New Model Cross-Validation Mean Score = 0.836
- New Model Standard Deviation = 0.003

Using statistical reasoning, determine if the new model is significantly better than the baseline.

Explain your reasoning briefly and conclude with either:

SIGNIFICANT IMPROVEMENT
or
NOT SIGNIFICANT

Your final answer must contain only one of the two options above.
"""
    response = llm.prompt(prompt)

    kbench.assertions.assert_in(
        "SIGNIFICANT IMPROVEMENT",
        response,
        expectation="The response should conclude with 'SIGNIFICANT IMPROVEMENT'."
    )
    kbench.assertions.assert_not_in(
        "NOT SIGNIFICANT",
        response,
        expectation="The response should not contain 'NOT SIGNIFICANT'."
    )

evaluate_model_improvement.run(kbench.llm)
```
