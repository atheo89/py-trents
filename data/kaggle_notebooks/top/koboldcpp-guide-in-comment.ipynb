## SnowDrop By TrashPanda 🐼
#### Performs well with reasoning and without it                       Doesn't hurt to try 😬
check out the link for recommend settings https://huggingface.co/trashpanda-org/QwQ-32B-Snowdrop-v0

###### OR replace the link below with the model of your choice 😂
Example URL = "https://huggingface.co/Hasnonname/Julleimm-Q6_K-GGUF/resolve/main/julleimm-q6_k.gguf?download=true"

Lower quants like Q3 for faster response speed or Q5 and above for quality and slow response speed.

```python
URL = "https://huggingface.co/bartowski/trashpanda-org_QwQ-32B-Snowdrop-v0-GGUF/resolve/main/trashpanda-org_QwQ-32B-Snowdrop-v0-Q4_K_L.gguf?download=true" #URL to the file of the model ex. "https://huggingface.co/MODEL/ID/blob/main/model.gguf"
#-----------------------------------------------------------------------------
from urllib.parse import urlparse, parse_qs

def extract_parts(url):
    parsed_url = urlparse(url)
    path_parts = parsed_url.path.strip('/').split('/')
    query = parse_qs(parsed_url.query)
    
    model_name = path_parts[0]+"/"+path_parts[1]
    version = path_parts[3]
    file_name = path_parts[4]
    
    return model_name, version, file_name

REPO_ID, REVISION, FILE = extract_parts(URL)

DIR = REPO_ID.replace("/", "_")

if REVISION != "" and REVISION != "main":
    DIR = f"{DIR}_{REVISION}"

print("Model Name: " + DIR)
print("Version: " + REVISION)
print("File: " + FILE)

from huggingface_hub import hf_hub_download

FULLPATH = f"/kaggle/models/{DIR}"

if REVISION != "":
    hf_hub_download(repo_id=REPO_ID, filename=FILE, local_dir_use_symlinks=False, revision=REVISION, local_dir=FULLPATH)
    
if REVISION == "":
    hf_hub_download(repo_id=REPO_ID, filename=FILE, local_dir_use_symlinks=False, local_dir=FULLPATH)
```
